```python
Mean = app_train['AMT_INCOME_TOTAL'].mean()
print ('Mean of AMT_INCOME_TOTAL is ' + str(Mean))
STD = app_train['AMT_INCOME_TOTAL'].std()
print ('STD of AMT_INCOME_TOTAL is ' + str(STD))
Variation = app_train['AMT_INCOME_TOTAL'].var()
print ('Variation of AMT_INCOME_TOTAL is ' + str(Variation))


app_train['AMT_INCOME_TOTAL'].hist(bins=20)
plt.xlabel('Interval')
plt.ylabel('Value')
plt.title('AMT_INCOME_TOTAL')

```
